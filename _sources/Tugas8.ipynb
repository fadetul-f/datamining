```python
from scipy.io import arff
import pandas as pd
```

```python
data = arff.loadarff('/content/drive/MyDrive/datamining/tugas/messidor_features.arff')
df = pd.DataFrame(data[0])
```

```python
col_names = []
for i in range(20):
    if i == 0:
        col_names.append('quality')
    if i == 1:
        col_names.append('prescreen')
    if i >= 2 and i <= 7:
        col_names.append('ma' + str(i))
    if i >= 8 and i <= 15:
        col_names.append('exudate' + str(i))
    if i == 16:
        col_names.append('euDist')
    if i == 17:
        col_names.append('diameter')
    if i == 18:
        col_names.append('amfm_class')
    if i == 19:
        col_names.append('label')
```

```python
df.columns = [col_names]
```

```python
df.head()
```

```python
y = df['label'].values
```

```python
x = df.drop(columns=['label'], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
print(scaler.fit(df))
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit (y)
```

```python
y=le.transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.naive_bayes import GaussianNB

training, test = train_test_split(df, train_size = 0.8, test_size = 0.2, shuffle=False)
training_label, test_label = train_test_split(y, train_size = 0.8, test_size = 0.2, shuffle=False)

clf2 = GaussianNB()
clf2.fit(training, training_label)

```

```python
from IPython.core.events import post_execute
from sklearn.metrics import accuracy_score
import numpy as np

probas = clf2.predict_proba(test)
probas = probas[:,1]
probas = np.round(probas)
pred = probas
akurasi = accuracy_score(test_label, pred)
akurasi
```
