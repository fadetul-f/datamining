$$Biner$$

---

atribut yang hanya memiliki 2 nilai mutlak, seperti 0 dan 1. 

Atribut binary dapat dibagi lagi dalam dua jenis yakni:
**teks tebal**
1.   Simetris: Kedua nilai sama derajat gunanya (misal: gender: laki-laki & perempuan).
2.   Asimetris: Kedua nilai tidak sama derajat gunanya (misal: hasil: lulus & gagal). Operasi yang dapat dilakukan distinctness.daftar


```python
#Menampilkan data biner dari dataset
X[["gender","test preparation course"]].head()
```

```python
#code gendernya
pria = "male"
lulus = "completed"

#nilai konversi biner
one = 1
zero = 0

def change_biner(gender):
  return one if gender == pria else zero
def change_binertes(test_preparation_course):
  return one if test_preparation_course == lulus else zero
```

```python
#Update Nilai dari Gender dan Test Preparation
X['gender'] = X['gender'].apply(change_biner)
X['test preparation course']= X['test preparation course'].apply(change_binertes)

X[["gender","test preparation course"]].head()
```

$$Perhitungan Biner(Simetris)$$

---



$$Rumus: d(i,j) = \ \frac{r+s} {q+r+s+t} $$



$$Keterangan: $$

                                                         *   r: i=1, j=0
                                                         *   s: i=0, j=1 
                                                         *   q: i=1, j=1
                                                         *   t: i=0, j=0 

```python
def binersimetris(i,j):
  q=r=s=t=0
  for col in ['gender','test preparation course']:
    if X[col][i]==1 and X[col][j] == 1:
      q+=1
      print("nilai q =",q)
    elif X[col][i]==1 and X[col][j]==0:
      r+=1
      print("nilai r =",r)
    elif X[col][i]==0 and X[col][j] == 1:
      s+=1
      print("nilai s =",s)
    else :
      t+=1
      print("nilai t =",t)
  print("Rumus = (r+s)/(q+r+s+t)")
  return (r+s)/(q+r+s+t)

print("Nilai Biner = ", binersimetris(1,3))
```

$$Nominal$$

---

Nominal
Nominal adalah jenis atribut yang hanya menyediakan nilai yang cukup untuk membedakan antara satu objek dengan objek lainnya. Nilai dari atribut nominal adalah nama benda atau dapat juga berupa simbol. Jenis atribut ini hanya bisa dilakukan operasi membandingkan sama atau tidaknya suatu data.

**Contoh :** nama orang, alamat, nomor ktp, kode pos, warna mata dan lain-lain.

Operasi yang bisa dilakukan adalah distinctness.


```python
X[["race/ethnicity","parental level of education"]].head()
```

$$Perhitungan Nominal$$

---

$$ Rumus: d(i,j) = \frac{m-p}{p}$$



$$Keterangan : $$

                          *   m = jumlah data yang dibandingkan
                          *   p = jumlah data sama

```python
def d1(i,j):
    P = 4
    M = 0
    for col in ["race/ethnicity","parental level of education"]:
        if X[col][i] == X[col][j]:
            M += 1

    return (P-M)/P

print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

$$Standar Deviasi$$

---
$$Rumus:𝔡=\sqrt{\frac{sigma(x_i-μ)^2}{N}}$$

$$z=\frac{x-μ}{𝔡}$$

```python
#normalisasi math score
std1 = 15.16308
mean1 = 66.08900
for col1 in ['math score']:
  data1 = X[col1]
  hasil1 = (data1-mean1)/std1
  print(hasil1)
#normalisasi reading_score
std2 = 14.600192
mean2 = 68.054000
for col2 in ['reading score']:
  data2 = X[col2]
  hasil2 = (data2-mean2)/std2
  print(hasil2)

#normalisasi writing_score
std3 = 15.195657
mean3 = 68.054000
for col3 in ['writing score']:
  data3 = X[col3]
  hasil3 = (data3-mean3)/std3
  print(hasil3)
```

$$Menghitung-Jarak-Numeric-dengan-Metode-Euclidean-Distance$$

---

$$Rumus:$$$$d(i,j)=\sqrt{(xᵢ₁-xⱼ₁)^2 + (xᵢ₂ - xⱼ₂)^2+.....}$$

```python
import math

def d2(x,y,h):
    result = 0
    for col in [hasil1, hasil2, hasil3]:
        result = math.pow(abs(col[x] - col[y]), h)
    return result**(1/h)


print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```

$$Similiarity$$

---

$$Rumus:$$$$Biner+Nominal+Numeric$$

```python
# Jarak d(1,2)
print("Jarak d(1,2) = ",d1(1,2)+d2(1,2,2)+binersimetris(1,2))

# Jarak d(1,3)
print("Jarak d(1,3) = ",d1(1,3)+d2(1,3,2)+binersimetris(1,3))

# Jarak d(1,4)
print("Jarak d(1,4) = ",d1(1,4)+d2(1,4,2)+binersimetris(1,4))
```
